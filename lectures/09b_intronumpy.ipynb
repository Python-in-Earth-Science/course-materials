## There's an easier way!! Use NumPy!!

````python
import numpy as np

# Convert data to NumPy arrays
peak_names = np.array(['Mount Mitchell', 'Mount Washington', 'Grandfather Mountain', 'Mount Rogers'])
initial_elevations = np.array([2037, 1916, 1800, 1746])
erosion_rates = np.array([2.7e-05, 3.1e-05, 4.2e-05, 3.8e-05])

# Calculate 4000 years of erosion in ONE operation
years = 4000
final_elevations = initial_elevations - (erosion_rates * years)
total_erosion = erosion_rates * years

# Analysis in two lines
tallest_peak = peak_names[np.argmax(final_elevations)]
most_eroded_peak = peak_names[np.argmax(total_erosion)]

print(f"Final elevations: {final_elevations}")
print(f"Tallest remaining: {tallest_peak}")
print(f"Most eroded: {most_eroded_peak}")
````

## What is an array?
In computer programming, an **array** is a structure for storing and retrieving data. Numpy uses `ndarray`s, which are similar to Python lists, but with some important differences:
- all objects in an `ndarray` have the same data type
    - default is `float`, but you can change this 
- an `ndarray` can have multiple dimensions (that's the 'n' in 'nd')
    - similar to nested lists, but with more flexibility
- mathematical operations are applied to *all* elements of the array at once.

We often talk about an array as if it were a grid in space, with each cell storing one element of the data. 

For instance, if each element of the data were a number, we might visualize a “one-dimensional” array like a list:<br>

#### Some additional rules for arrays:
- All elements of the array must be of the same type of data.
- Once created, the total size of the array can’t change.
- The shape must be “rectangular”, not “jagged”; e.g., each row of a two-dimensional array must have the same number of columns.

## Indexing Arrays

Arrays can be indexed using an extended slicing syntax, similar to what we did with lists.

#### Basic Indexing
We can do basic indexing the same way we did it for lists. Just remember that we use zero indexing! For example:

````python
# Create a sample array
elevations = np.array([1200, 1350, 1500, 1650, 1800])

# Access individual elements (remember: Python starts counting at 0!)
print(elevations[0])    # First element: 1200
print(elevations[2])    # Third element: 1500  
print(elevations[-1])   # Last element: 1800
print(elevations[-2])   # Second-to-last: 1650
````

#### Slicing
We can also slice the same way we did with lists. This is super handy.

````python
# Get a range of values using [start:stop]
print(elevations[1:4])   # Elements 1, 2, 3 (NOT 4): [1350 1500 1650]
print(elevations[:3])    # First 3 elements: [1200 1350 1500]
print(elevations[2:])    # From element 2 to end: [1500 1650 1800]
print(elevations[::2])   # Every 2nd element: [1200 1500 1800]
````

## Operations on arrays: two arrays

**When the two `ndarrays` are the same size**:
- the operation is done on each pair of corresponding values
- a new array is produced<br>

````python
a1 = np.array([1, 2, 3, 4])
a2 = np.array([5, 6, 7, 8])
a3 = a1 + a2
print(a3)
print(type(a3))
````

**When the two `ndarrays` are DIFFERENT sizes**:
- NumPy tries to "broadcast" the arrays together
- broadcasting works when the arrays can be made compatible by stretching dimensions
- broadcasing fails when the dimensions are incompatible, giving you an error

**Two dimensions are compatible if:**
- they're equal<br>
OR
- one of them has a dimension of size 1
    - `np.array([4])` has a dimension of size 1.
    - `np.array([3, 2])` has dimension of size 2.

Here's an example of ***incompatible arrays***:

````python
thing1 = np.array([1,2,3])
thing2 = np.array([4,5])
thing1 + thing2
````

Numpy compares shapes by **aligning them from the right side and checking to see if each dimension is compatible**.
First, let's determine the shape of `thing1` and `thing2`.

````python
print(thing1.shape)
print(thing2.shape)
````

In other words, `thing1` has a dimension of 3, while `thing2` has a dimension of 2. Aligning these shapes from the right, we can see that their only dimension is not equal and neither has a dimension of 1, so they're incompatible!

Another way to visualize this:<br>
```
thing1: (3,)
thing2: (2,)
         ↑
         └── only dimension: 3 vs 2 ✗ (neither is 1, and they're different)
```

Here's an example of ***compatible arrays***:

````python
# We recorded daily temperatures at 4 weather stations over 3 days
temperatures = np.array([[20, 22, 18, 25],
                         [19, 24, 17, 26],
                         [21, 23, 19, 24]])

# we know that our weather stations are off by the same amounts all the time, but it varies by station like this:
elevation_correction = np.array([-2, -2, -3, 0])

print("Temperatures shape:", temperatures.shape)
print("Elevation correction shape:", elevation_correction.shape)
````

Because `elevation_correction` is (4,), NumPy will treat this as (1,4) (1 row, 4 columns). If we were to align our array sizes from the right we would see:
```
(3,4)
(1,4)
```
Because they both have 4 in their right-most measurement (i.e., they both have 4 columns), we can *stretch* the `elevation_correction` array from (1,4) to (3,4).

Another way to visulaize it:
```
temperatures:         (3, 4)
elevation_correction: (1, 4)
                       ↑  ↑
                       |  └── Last dimension: 4 vs 4 ✓ (same)
                       └────── Second-to-last: 3 vs 1 ✓ (one is 1, so stretches)
```

<center>If we wanted to <b>add these two arrays together</b>, we could!</center><br>

Numpy would simply do the following:
- take the first row of `temperatures` and add to it the value in the corresponding position in `elevation_correction`
```
    20 - 2 = 18
    22 - 2 = 20
    18 - 3 = 15
    25 - 0 = 25
```
- then it would do the same to the second row:
```
    19 - 2 = 17
    24 - 2 = 22
    17 - 3 = 14
    26 - 0 = 26
```

#### What would the values of the third row be after the addition of these two arrays?

[19, 21, 16, 24]

#### Let's do it now!

````python
corrected_temps = temperatures + elevation_correction
print("Result shape:", corrected_temps.shape)
print("Corrected temperatures:\n", corrected_temps)
````

**NOTE**: You can use
`+`, `-`, `*`, `/`, `**` or any other operation between compatible arrays. Broadcasting will work the same way for all of them!

### Sometimes, our arrays might have the right numbers, but the wrong shape for the calculation you want.
For example, what if we have the array `data` and we want to multiply each row by a different value?

````python
data = np.array([[10, 20, 30],    # Row 1
                 [40, 50, 60]])   # Row 2

multipliers = np.array([[2, 3]])  # shape of that: (1, 2) - our intention is to multiply each row in data by a value here
````

As it is, this won't work how you want it to! If were to were to do:

````python
data * multipliers
````

````python
data.shape
````

````python
multipliers.shape
````

You see that we get an error because it's trying to broadcast something that's incompatible:
```
(2,3)
(1,2)
```

**BUT there are two values stored in `multipliers`! And two rows!**

### How do we need to change `multipliers` in order for these two arrays to be compatible?**

<center> We can flip the dimensions from (1,2) to (2,1) and all of a sudden these arrays are compatible! </center>

#### <center>This is called <b>transposing</b> the data.</center>

**Transposing** is a process that switches the rows and column indices. So in our case, we need the number of columns to become the number of rows. This could also look like:



To transpose something, we can use the `.T` method.

````python
multipliers_T = multipliers.T
result = data*multipliers_T
print(result)
````

**NOTE**: In this example, we've used the "list within list" syntax and that worked! Transpose flips dimensions, so you have to have at least 2 dimensions for it to do anything meaningful.
- **1D arrays** have only one dimension, so `.T` has no effect
- **2D+ arrays** have multiple dimensions so they can be flipped.

#### What do you do if you have a 1D array that you'd like to flip from a row to a column?

First, you'll need to use `.reshape()` it to 2D and *then* transpose it:

````python
arr_1d = np.array([1, 2, 3])           # (3,)
arr_2d = arr_1d.reshape(1, 3)          # (1, 3)
# or in one step: arr_1d.reshape(1, 3)
arr_transposed = arr_2d.T              # (3, 1)
print("Original array:", arr_1d)
print("Reshaped array:", arr_2d)
print("Transposed array:\n", arr_transposed)
````
