## Arithmetic Operator Precedence


Operator precedence defines the order in which operations are performed. This is logical--no tricks, I swear!
Here's what you need to know:

1. Parentheses have the highest precedence, meaning that expressions inside parentheses are evaluated first.

2. Exponation has precedence over multiplication

3. Multiplication * has higher precedence than addition + and subtraction, and therefore multiplications are evaluated before additions.

4. Multiplication is at the same *precendence level* as division, and the statement is executed left to right.
This is the same for addition and subtraction.

This then gives you, the total solar power delivered to Earth

$$P_{in} = S_0 (1 - a) \pi r_e^2$$
<center> or </center>
$$P_{in} = S_0 (1 - a) A_c$$

### Okay, now that we have the incoming solar radiation, we now need to know how  much power Earth is emitting.

<span style="color: purple;">Earth's radiation (outgoing radiation) <span> <br>

According to the ***Stefan-Boltzmann Law***, radiant power emmitted by a blackbody per unit surface area depends on temperature:

$$M = \sigma T^4$$

- $M$ is the power radiated in W/m$^2$
- $\sigma$ is the Stefan-Boltzmann constant in W m$^{-2}$ $^\circ$K$^{-4}$
    - $\sigma \approx 5.67\times 10^{-8}$ (how to remember: 5-6-7-8)
- $T$ is the temperature of the body in degrees Kelvin

When we calculated the incoming radiation, we used the cross-sectional area of Earth because at any given time, that is what is absorbing radiation from the sun. For outgoing radiation, the entire Earth is radiating, so we need to now multiply by the surface area of Earth instead. This equation then becomes:
$$M = 4 \pi r^{2} \sigma T^{4}$$

# <span style="color: purple;">REMEMBER OUR MOTIVATION: we want to better understand what sets Earth's average surface temperature ($T$)<span>
