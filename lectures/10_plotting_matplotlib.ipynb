### Important: `plt.show()`
- `plt.show()` displays your plot in a window
- in Jupyter notebooks, plots often display automatically, but it's good practice to include it because that's not always the case in other IDEs
- after `plt.show()`, the figure is cleared and you start fresh for the next plot
### Starting Fresh
Each time you want to create a new, separate plot, matplotlib needs to know. You can tell it you're creating a new figure in one of two ways:
1. Use `plt.show()` - displays the current plot and clears it for the next one.
2. Use `plt.figure()` - explicitly creates a new figure

### Controlling Figure Size
By default, matplotlib creates figures that are 6.4 inches wide by 4.8 inches tall. You can change this by using the `figsize` parameter.

````python
# default size
plt.plot([1,2,3],[4,5,6])
plt.title('Default Size Figure')
plt.show()

# custom size: (width, height) in inches
plt.figure(figsize=(10,6)) # 10 inches wide, 6 inches tall
plt.plot([1,2,3],[4,5,6])
plt.title('Larger Figure')
plt.show()
````

#### When to adjust figure size
- Make it **larger** when you have lots of data or need to see details.
- Make it **wider** for time series data (like our temperature cycles)
- Make it **square** for scatter plots where x and y should have equal visual weight
- Adjust for **subplots** (will learn about later) ; more plots usually need more space

### Other Useful Figure Settings
**Setting DPI** (resolution): DPI = 'dots per inch'
- higher DP = sharper image
```

```
**Tight layout** (preventing label cutoff): sometimes labels or titles get cut off at the edges. `tight layout()` fixes this automatically.
```

plt.show()
```

## Two Ways to Create Plots: `plt` vs `ax`
You'll see two different styles when looking at matplotlib code online. They both work, but it's important to understand the difference.
### **Style 1**: The `plt` approach (what we've used so far)
Use this for simple plots, quick visualization, and when you have a single plot.

````python
plt.plot(x, y)
plt.xlabel("X Label")
plt.title("My Plot")
plt.show()
````

### **Style 2**: The `fig, ax` approach (object-oriented)
Use this when you have multiple plots, when you want more control, or when you want to look/feel more professional.

````python
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel("X Label")
ax.set_title("My Plot")
plt.show()
````

#### Earlier, I told you that `plt.plot` automatically creates a figure and an axes for you. `plt.subplots` separates those into two different outputs.  

#### `plt.subplots()` returns a tuple `(figure_object, axes_object)`.
#### `fig, ax = plt.subplots()` then unpacks that tuple into two separate variables: `fig` and `ax`

### Why the difference?
- `plt` approach: matplotlib manages everything automatically behind the scenes
- `fig, ax` approach: you explicitly control the figure and axes objects
    - if you're looking for ways to plot things online, you'll often see the `fig, ax` approach because it's more flexible and considered better practice
    - this is also how you would control a plot with multiple axes

**Note**: `plt.xlabel()` becomes `ax.set_xlabel()` and `plt.title()` becomes `ax.set_title()`

## Understanding **Subplots**
Even if you only want ONE plot, you'll often see code like this:
```
fig, ax = plt.subplots()

```
This will create 1 figure with 1 axis.<br>

This is the **preferred modern approach** because:
1. It's explicit about creating the figure and axes
2. It's consistent whether you have 1 plot or many plots
3. It gives you more control over your plot

**The notation: `fig, ax`**
- `fig` = the figure (the entire window/canvas)
- `ax` = the axes (the single plot area)
- You'll see `ax` (singular) for one plot and `axes` (plural) for multiple plots

### **Subplots**: Creating a single plot with the `fig, ax` approach

````python
# create a figure and axes
fig, ax = plt.subplots(figsize=(8,6))

# plot on the axes object
ax.plot(month_numbers, temperatures[0], 'b-o')
ax.set_xlabel('Month')
ax.set_ylabel('Temperature (C)')
ax.set_title('Arctic Station Temperature Cycle')
ax.grid(True)

plt.show()
````

### **Subplots**: Multiple Plots in One Figure with the `fig, ax` approach
When you want multiple plots, `plt.subplots()` creates them all at once.<br>
To create a 2x2 grid of subplots you do:

````python
fig, axes = plt.subplots(2,2, figsize=(12,8))
````

As you can see, this creates 1 figure with 4 axes arranged in 2 rows x 2 columns. <br>

`axes` is now a 2D array: `axes[row, column]`. You can acces each individual plot the same way to how we indexed lists or arrays. So:
- axes[0,0] = upper left corner plot
- axes[0,1] = upper right corner plot

#### How would we refer to the lower two plots?

axes[1,0] = lower left corner plot
axes[1,1] = lower right corner plot

#### We can access and change individual plots using those indices.

In 1993, after five years of drilling, the Greenland Ice Sheet Project Two (GISP2) penetrated through the icesheet and 1.55 meters into bedrock, recovering an ice core 3053 meters deep.

One crucial analysis is literally **counting annual layers**. Each year leaves a distinct layer** in the ice. By measuring layer thickness, scientists can:
- count years backward in time
- identify periods of high vs low snowfall
- detect climate chagnes that affected accumulation rates

**note that these layers compress over time, but we will be ignoring that for now.<br>
