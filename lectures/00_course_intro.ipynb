## Why Python?

This course teaches Python — one of the most widely used languages in scientific computing. A few reasons we use it:

- **Readable**: Python code is close to plain English, which makes it easier to learn and to share with collaborators
- **Powerful**: the scientific Python ecosystem (NumPy, pandas, Matplotlib, and more) covers almost everything you'd need for data analysis, visualization, and modeling
- **Everywhere**: Python is used across Earth science, climate modeling, hydrology, geophysics, remote sensing, and beyond — knowing it opens a lot of doors
- **Free and open source**: no licenses, no cost, runs on any machine

The goal of this course is not to make you a software engineer — it is to make you a scientist who can write code to answer scientific questions.

## $\LaTeX$ math in notebooks

In Markdown cells we can write properly typeset math using $\LaTeX$ (pronounced "lay-tek"). You'll use this a lot in this course — any time you want to document what equation your code is computing.

There are two modes:

**Inline math** — wrap with single dollar signs, renders within a sentence:  
`The slope angle $\theta$ is measured in degrees.` → The slope angle $\theta$ is measured in degrees.

**Display math** — wrap with double dollar signs, renders on its own line:  
```
$$Q = A \cdot v$$
```
$$Q = A \cdot v$$

#### Common $\LaTeX$ symbols you'll use in this course

| What you type | What you get | Name |
|---|:---:|---|
| `$\pi$` | $\pi$ | pi |
| `$\theta$` | $\theta$ | theta |
| `$\alpha$` | $\alpha$ | alpha |
| `$x^{2}$` | $x^{2}$ | superscript |
| `$x_{i}$` | $x_{i}$ | subscript |
| `$\frac{x}{y}$` | $\frac{x}{y}$ | fraction |
| `$\sqrt{x}$` | $\sqrt{x}$ | square root |
| `$\approx$` | $\approx$ | approximately equal |
| `$\Delta$` | $\Delta$ | delta (change in) |

## Code cells

Code cells contain Python. When you run one, the kernel executes the code and displays any output below the cell.

A few things to know:
- Lines starting with `#` are **comments** — they are ignored by Python but essential for human readers
- `print()` outputs text to the screen
- Variables and results from one cell are available in all cells below it (as long as you've run the cells in order)

Here's a simple example — a calculation you might actually do in your research:

````python
# Estimating the volume of a landslide deposit
# Data from field measurements

length = 450      # meters
width = 120       # meters
avg_depth = 3.5   # meters

volume = length * width * avg_depth

print("Estimated deposit volume:", volume, "cubic meters")
print("That's equivalent to", round(volume / 2400, 1), "Olympic swimming pools.")
````
