#### `np.full()`

`np.full()` is the string equivalent of `np.zeros()`. Some hints for use:
- `np.full(shape, fill_value, dtype)`
- shape should be in the format of (rows, cols)
- setting fill_value equal to `""` will create empty strings ready to be filled
- the different dtypes for strings specify the size of the string
    - `U5` = max 5 characters
    - `U10` = max 10 characters
    - `U15` = max 15 characters

#### `np.where()`
Earlier, we learned about `if-elif-else` statements. When working with NumPy arrays, you can also nest `np.where()` to accomplish similar things.

```python
test = np.array([4,5,2,7,8,4,2])
#For two categories:
two_cats = np.where(test<2, "Bloop", "Not Bloop")
#For three categories:
three_cats = np.where(test<2, "Bloop", 
                      np.where(test<5, "Blop", "Not Bloop or Blop"))
#For four categories:
four_cats = np.where(test<2, "Bloop",
                     np.where(test<5, "Blop",
                              np.where(test<8, "Bleep", "Sleep")))
print(two_cats)
print(three_cats)
print(four_cats)
```

If you're trying to meet two conditions:

```python
# two conditions
np.where((test>2) & (test<5), "Blop", "Not Bloop or Blop")
```
