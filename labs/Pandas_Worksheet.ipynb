## **Exercise 1: Working with Functions**
### **Part A: Using a Given Function**
Here's a complete function that calculates summary statistics for discharge data. **Read it carefully** and run it.

````python
def calculate_flow_stats(dataframe, column_name='discharge_cfs'):
    """
    Calculate summary statistics for discharge data.

    Parameters:
    -----------
    dataframe: DataFrame
        Discharge data
    column_name: str
        Name of discharge column (default: 'discharge_cfs')

    Returns:
    --------
    dict : Dictionary with mean, min, max, and std values
    """
    stats = {
        'mean': dataframe[column_name].mean(),
        'min': dataframe[column_name].min(),
        'max': dataframe[column_name].max(),
        'std': dataframe[column_name].std()
    }
    return stats

#Test the function
flow_stats = calculate_flow_stats(df)
print(f"Mean discharge: {flow_stats['mean']:.2f} cfs")
print(f"Min discharge: {flow_stats['min']:.2f} cfs")
print(f"Max discharge: {flow_stats['max']:.2f} cfs")
print(f"Standard deviation: {flow_stats['std']:.2f} cfs")
````
