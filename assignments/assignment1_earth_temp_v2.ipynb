### Exhaustive Answer:

To calculate the Earth's average radiating temperature, we first had to derive a few other equations.
First, we derived an equation for the cross-sectional area of the Earth, $A_c = \pi(r_e)^2$ , which gives the cross-sectional area in square meters. This equation has variables: 
    radius of the Earth: $r_e$ = 6378 meters

We also had to calculate the total incoming solar power to Earth, using $P_{in} = S_0*(1-a)*A_c$ , which gives the incoming solar power in W/$m^2$. This equation has variables:
    Solar constant: $S_0$ = 1360 W/m,
    Albedo of the Earth: $a$ = 0.31,
    Cross-sectional area of the Earth: $A_c = \pi(r_e)^2$ = 127731695.76 square meters

Next, we calculated the Earth's energy output using the Stefan-Blotzmann Law. $$M=4\pi*r^2*\sigma*T^4$$
This equation has variables:
    Stefan-Boltzmann constant: $\sigma \approx 5.67\times 10^{-8}$ in W m$^{-2}$ K$^{-4}$
    Temperature of the body: $T$ = today's temperature in Kelvin

Finally, we set the outgoing radiation equal to the ingoing radiation: $$4 \pi r^2 \sigma T^4 = \pi r^2 S_0 (1-a)$$
and solved for $T$ $$T = \sqrt[4]{\frac{S_0 (1-a)}{4\sigma}}$$

This equation can be used to calculate the Earth's average radiating temperature.
